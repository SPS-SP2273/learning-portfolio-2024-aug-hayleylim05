### Task 1 ###

```python
dict_months = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

```

```python
file = 'oct-08_singapore.txt'

month = file[:3] # taking the first three letters (which is the month)
date_city = file[3:] # taking everything that comes after the month
number = dict_months[month] # used the month as an index in the dictionary to get the number
new_name = str(number+date_city)
print(new_name)
```

### Task 2 ###

```python
def rename_my_file(old_file_name):
    month = old_file_name[:3]
    date_city = old_file_name[3:]
    number = dict_months[month]
    return str(number+date_city)
    
rename_my_file('oct-08_singapore.txt')
```

### Task 3 ###

```python
file_list = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

for file in file_list:
    new_file_name = rename_my_file(file)
    print(file, '---->', new_file_name)
```

### Task 4 ###

```python
import os
import glob
import shutil



files = glob.glob('data_numbers/**.txt') # finding the files
try:
    for file in files:
        old_file_name = file[13:] # doing this to keep the part before the month (aka 'data_numbers/')
        new_file_name = rename_my_file(old_file_name)
        file_name = file[:13]+new_file_name
        os.rename(file, file_name)

except:
    print("Files are already renamed")
```

### Task 5 ###

```python
cities_folders = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']


try:
    os.mkdir('cities') # create folder called cities in current working directory
    for city in cities_folders:
        path = os.path.join('cities', city) # creating a path to create new folders (the cities_folders)within the folder 'cites'
        print(f'Creating {path}')
        os.mkdir(path)
        
except:
    print('Files are created')
```

### Task 6 ###

```python
files_cairo = glob.glob('data_numbers/**cairo*') # finding all files that contai "cairo"

for file in files_cairo:
    new_path = os.path.join('cities', 'Cairo')
    shutil.copy(file, new_path) # copying file to 'Cairo' folder
```

### Task 7 ### 

```python
cities = ['cairo', 'dhaka', 'jakarta', 'karachi', 'manila',
 'melbourne', 'osaka', 'shanghai', 'singapore', 'tokyo']

for city in cities:
    files_of_interest = glob.glob('data_numbers/**'+city+'*') # finding all the files of that city
    for city_folder in cities_folders:
        if city[1:] == city_folder[1:]:
            folder_of_interest = str(city_folder) # identifying the right folder based on the city of the file
    for file in files_of_interest:
        new_path = os.path.join('cities', folder_of_interest)
        shutil.copy(file, new_path)
        print(f'Copied file to {new_path}')
        


        
```
