```python
from matplotlib import pyplot as plt
```

### Exercise 1 ###

```python
# Some data for plotting

x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='maroon', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

### Exercise 2 ###

```python
import numpy as np

x = np.arange(0, 51, 0.1) # to make it less choppy as opposed to 'x = np.array(range(50))'
y = np.exp(-x/10)*np.sin(x)

plt.plot(x, y, color='blue')
plt.ylim(-1,1)
plt.xlabel('x', color='darkorange', size='20')
plt.ylabel('y', color='darkorange', size='20')
plt.grid(alpha=0.25) # 25% opacity
plt.title(r'$y=e^{-x/10} \sin(x)$', size='20')

```
